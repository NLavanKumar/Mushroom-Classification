##### data set Categorical data
##### we have to convert categorical  data into numerical data
##### so we use Label Encoder
##### Label Encoding is a technique used to convert categorical data into numerical data by assigning a unique numerical label to each category within a feature.

##### To use label encoding the data set must be in categorical

```python
data.info()
```

##### by default a non-numerical column is representes by 'object datatype'

```python
#coverting our columns datatype to Category
data = data.astype('category')
```

```python
data.dtypes
```

```python
#Label Encoding is a technique used to convert categorical data into numerical data by assigning a unique numerical label to each category within a feature.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in data.columns:
    data[col] = le.fit_transform(data[col])
```

### 8. Store Feature Matrix in X and Response(Target)in Vector y

```python
selected_features = [
    'gill-size',
    'gill-color',
    'bruises',
    'ring-type',
    'gill-spacing',
    'odor',
    'population',
    'habitat',
    'cap-surface',
    'spore-print-color'
]

# Define X and y
X = data[selected_features]
y = data['class']
```

```python
X
```

```python
y
```

### 9. Applying PCA

```python
# Principal Component Analysis (PCA)
# High-dimensional data can be computationally expensive and difficult to visualize.
# PCA helps reduce the number of features
# This can improve the performance and efficiency of your machine learning models.
```

```python
from sklearn.decomposition import PCA
pca1=PCA(n_components=10)
pca1_fit=pca1.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
import joblib

# Step 1: Initialize PCA
pca = PCA(n_components=10)

# Step 2: Fit PCA on your dataset (X must be numeric and preprocessed)
X_pca = pca.fit_transform(X)

# Step 3: Save the fitted PCA to 'pca1.pkl'
joblib.dump(pca, 'pca1.pkl')

print("PCA saved successfully as 'pca1.pkl'")
```

```python
pca1.explained_variance_ratio_ #. It represents the proportion of variance explained by each of the selected principal components.
```

```python
#explained_variance_ratio_ helps you determine the optimal number of principal components to retain for dimensionality reduction.
#You can choose the number of components that explain a sufficient amount of variance (e.g., 95% or 99%).
#This allows you to reduce the complexity of your data while preserving most of the important information.
```

```python
sum(pca1.explained_variance_ratio_ ) # n components retaing ---------%(information)
```

### 10. Splitting The Dataset into Training Set And Test Set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(pca1_fit,y,test_size=0.2,random_state=42)
```

```python
print('X_train',X_train.shape)
print('X_test',X_test.shape)
print('y_train',y_train.shape)
print('y_test',y_test.shape)
```

```python
data.head()
```

```python
# our  target value is categrical so we have use classification models
#Classification in machine learning is the process of predicting the class or category of a given data point
```

### 11. Import the models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

### 12. Model Training

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

rf = RandomForestClassifier()
rf.fit(X_train, y_train)

svc = SVC()
svc.fit(X_train, y_train)

gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)
```

### 13. Prediction on Test Data

```python
y_pred1 = lr.predict(X_test)
y_pred2 = knn.predict(X_test)
y_pred3 = svc.predict(X_test)
y_pred4 = dt.predict(X_test)
y_pred5 = rf.predict(X_test)
y_pred6 = gb.predict(X_test)
```

### 14. Evaluating the Algoithm

```python
from sklearn.metrics import accuracy_score
```

```python
print("Acc Logistic Regression:",accuracy_score(y_test,y_pred1))
print("Acc KNN",accuracy_score(y_test,y_pred2))
print("Acc Decision Tree",accuracy_score(y_test,y_pred4))
print("Acc Random Forest",accuracy_score(y_test,y_pred5))
print("Acc SVC",accuracy_score(y_test,y_pred3))
print("Acc Gradient Boosting",accuracy_score(y_test,y_pred6))
```

```python
final_data=pd.DataFrame({'Models':['LR','KNN','SVC','DT','RM','GBC'],
              'ACC':[accuracy_score(y_test,y_pred1)*100,
                     accuracy_score(y_test,y_pred2)*100,
                     accuracy_score(y_test,y_pred3)*100,
                     accuracy_score(y_test,y_pred4)*100,
                     accuracy_score(y_test,y_pred5)*100,
                     accuracy_score(y_test,y_pred6)*100,]})
```

```python
from tkinter import *
import joblib
from tkinter import messagebox

def show_entry_fields():
    try:
        inputs = [int(entry.get()) for entry in entries]
        model = joblib.load('Mushroom_prediction')
        pca = joblib.load('pca1.pkl')
        result = model.predict(pca.transform([inputs]))

        result_label.config(
            text="🍄 This mushroom is Edible! ✅🌱" if result[0] == 0 else "☠️ Warning! This mushroom is Poisonous! ❌😵‍💫",
            fg="#0077B6" if result[0] == 0 else "#D00000",
            font=('Arial', 12, 'bold')
        )
    except Exception as e:
        messagebox.showerror("Error", f"Something went wrong:\n{e}")

# GUI setup
master = Tk()
master.title("Mushroom Classification Using Machine Learning")
master.configure(bg="#FFF8E7")  # Soft Cream background

Label(master, text="🍄 Mushroom Classification🍄", 
      bg="#2E8B57", fg="white", font=('Helvetica', 16, 'bold')).grid(row=0, columnspan=2, pady=10)

labels = [
    ("cap-shape", "bell=0, conical=1, convex=5, flat=2, knobbed=3, sunken=4"),
    ("cap-surface", "fibrous=0, grooves=1, scaly=3, smooth=2"),
    ("cap-color", "brown=4, buff=0, cinnamon=1, gray=3, pink=5, purple=6, red=2, white=7, yellow=8"),
    ("bruises", "bruises=1, no=0"),
    ("odor", "almond=0, anise=3, creosote=1, fishy=8, foul=2, musty=4, none=5, pungent=6, spicy=7"),
    ("gill-attachment", "attached=0, descending=1, free=2, notched=3"),
    ("gill-spacing", "close=0, crowded=2, distant=1"),
    ("gill-size", "broad=0, narrow=1"),
    ("gill-color", "black=4, brown=5, buff=0, chocolate=3, gray=2, green=8, orange=6, pink=7, red=1, white=10, yellow=11"),
    ("stalk-shape", "enlarging=0, tapering=1")
]

entries = []
for i, (feature, desc) in enumerate(labels):
    Label(master, text=f"{feature}:", font=('Arial', 10, 'bold'), bg="#FFF8E7", fg="#1D3557").grid(row=i+1, column=0, sticky=W, padx=10, pady=2)
    Label(master, text=desc, font=('Arial', 10), bg="#FFF8E7", fg="black").grid(row=i+1, column=1, sticky=W, padx=10, pady=2)
    entry = Entry(master, font=('Arial', 10), bg="white", fg="black")
    entry.grid(row=i+1, column=2, pady=2, padx=10)
    entries.append(entry)

Button(master, text='Predict 🍄', command=show_entry_fields, bg="#4CAF50", fg="white",
       font=('Arial', 12, 'bold'), padx=10, pady=5).grid(row=len(labels)+1, columnspan=3, pady=10)

result_label = Label(master, text="", bg="#FFF8E7", font=('Arial', 12))
result_label.grid(row=len(labels)+2, columnspan=3, pady=5)

mainloop()
```
